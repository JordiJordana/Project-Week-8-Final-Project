## Libraries to import

```python
import pandas as pd
```

## Read from the Excels that have de data

```python
nyt_1701 = pd.read_excel('../1. Data/NYT_201701.xlsx')
nyt_1702 = pd.read_excel('../1. Data/NYT_201702.xlsx')
nyt_1703 = pd.read_excel('../1. Data/NYT_201703.xlsx')
nyt_1704 = pd.read_excel('../1. Data/NYT_201704.xlsx')
nyt_1801 = pd.read_excel('../1. Data/NYT_201801.xlsx')
nyt_1802 = pd.read_excel('../1. Data/NYT_201802.xlsx')
nyt_1803 = pd.read_excel('../1. Data/NYT_201803.xlsx')
nyt_1804 = pd.read_excel('../1. Data/NYT_201804.xlsx')
```

## Concatenate all the data

Concatenate the data to manage it in one dataframe.

```python
nyt = pd.concat([nyt_1701,nyt_1702,nyt_1703,nyt_1704,nyt_1801,nyt_1802,nyt_1803,nyt_1804], axis=0, ignore_index=True)
nyt.tail()
```

## Drop unnecessary columns

Columns with no useful information for the Sentiment Analysis are droped.

```python
nyt.drop(columns=['abstract', 'articleID', 'byline', 'documentType', 'multimedia', 'newDesk', 'printPage', 'pubDate', 'sectionName', 'source', 'typeOfMaterial','webURL'], inplace=True)
nyt.head()
```

Keywords column is also droped. Since it conveys information, it needs some more explanation. The main reason why this column is dropped is that this words have intend no meaningful syntactic information.

```python
nyt.drop(columns=['keywords'], inplace=True)
nyt.head()
```

```python
nyt[nyt.headline == 'Unknown']
```

## Drop of rows with healdine as 'Unknown'

```python
nyt = nyt[nyt.headline != 'Unknown']
```
